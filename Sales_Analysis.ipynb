```python
# Import python libraries
# why we have used '%matplotlib inline'?
# Numpy is used to work on arrays.
# Pandas is used to work on DataFrame (Table).
# Matplotlib and Seaborn are used for data visualization.
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

# Import csv file
# encoding = 'unicode_escape' is used to avoid 'Encoding Error'.
a = pd.read_csv("Diwali_Sales_Data.csv", encoding = "unicode_escape")

# Number of rows and columns in this table?
print(a.shape)
print("\n")

# Show only topmost 5 rows
print(a.head(5))
print("\n")
```

```python
# The 'info()' method prints information about the DataFrame.
# Return a summary of the DataFrame.
# The information contains the number of columns, column labels, column data types, memory usage, range index, and the number of cells in each column (non-null values)
# The 'info()' method does not return any value, it prints the information.
a.info()
```

```python
# Drop blank columns 'Status' and 'unnamed1'
# axis = 1 matalab 
# inplace = True matalab 
a.drop(["Unnamed_01", "Unnamed_02"], axis = 1, inplace = True)
```

```python
# Check once whether those two columns have disappeared or not
print(a.info())
print("\n")
```

```python
# If 'Null', then 'True'....otherwise 'False'
print(pd.isnull(a))
```

```python
# How many 'Null' in every column?
print(pd.isnull(a).sum())
```

```python
# Check once more how many rows and columns are present in this DataFrame?
print(a.shape)
```

```python
# Drop null values
a.dropna(inplace = True)
```

```python
print(pd.isnull(a).sum())
```

# Exploratory Data Analysis

### Gender

```python
# How many orders from which 'Gender'?
# Make a count plot to show this
p = sns.countplot(data = a, x = "Gender", palette = "bone")

# Give a label above each bar....such as 7842 and 3409
for apanaApanaBar in p.containers:
    p.bar_label(apanaApanaBar)

# Set the dimensions of this figure as per your convenience....width and height
sns.set(rc = {"figure.figsize" : (17, 6)})

#
plt.show()
```

```python
# More orders have come from female customers thanfrom male customers
# 'Order' ----> Number of Order
```

```python
# Take a good look at 'groupby()'
# 
kisGenderSeItanaRupyaKaOrder = a.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

#
print(kisGenderSeItanaRupyaKaOrder)
print("\n")
```

```python
# 
kisGenderSeItanaRupyaKaOrder = a.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = True)

#
print(kisGenderSeItanaRupyaKaOrder)
print("\n")
```

```python
# Matalab female customers se zyyada rupya ka order aaya
# 
# Make a bar plot of this
```

```python
# Make a bar plot showing 'Gender' vs 'Amount'
# 
kisGenderSeItanaRupyaKaOrder = a.groupby(["Gender"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

# 
q = sns.barplot(data = kisGenderSeItanaRupyaKaOrder, x = "Gender", y = "Amount", palette = "icefire")

# Give a label above each bar
for apanaApanaBar in q.containers:
    q.bar_label(apanaApanaBar)

# Set the dimensions of this figure as per your convenience....width and height
sns.set(rc = {"figure.figsize" : (17, 6)})

# 
plt.show()
```

```python
# See how many orders came from female and how many orders came from male within a particular age group.
```

```python
# Now show this through a bar plot.
```

```python
# Make a bar plot showing 'Age_Group' vs 'Amount'.
# Matalab kee kis Age_Group ne kitana rupaya ka order kiya hai
kisAgeGroupSeKitanaRupayaKaOrder = a.groupby(["Age_Group"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

# 
t = sns.barplot(data = kisAgeGroupSeKitanaRupayaKaOrder, x = "Age_Group", y = "Amount", palette = "Greens")

# Give a label above each bar
for apanaApanaBar in t.containers:
    t.bar_label(apanaApanaBar)

# Set the dimensions of this figure as per your convenience....width and height
sns.set(rc = {"figure.figsize" : (17, 6)})

# 
plt.show()
```

```python
# Now show this through a bar plot.
```

```python
# Kis state se kitana rupaya ka order aaya?
# Show only topmost 10 states.
# kya 'Amount' ke liye bhee wahee order aaya jo 'Order' ke liye aaya tha?....nahee
kisStateSeKitanaRupayaKaOrder = a.groupby(["State"], as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False).head(10)

#
v = sns.barplot(data = kisStateSeKitanaRupayaKaOrder, x = "State", y= "Amount", palette = "Oranges")

# Give a label above each bar
for apanaApanaBar in v.containers:
    v.bar_label(apanaApanaBar)

# Set the dimensions of this figure as per your convenience....width and height
sns.set(rc = {"figure.figsize" : (17, 6)})

#
plt.show()
```
